```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
import numpy as np
f= open("q6.txt","r")
z=f.read().split("\n")
del z[-1]
ylist = [int(str.split("\t")[1]) for str in z]
xlist = list(range(1,577))

```

```python
for i in range(4):
    xtrain = xlist [i*144:(i+1)*144]
    ytrain = ylist [i*144:(i+1)*144]
    xbar = np.mean(xtrain)
    ybar = np.mean(ytrain)

    b1 = sum(np.multiply(np.subtract(xtrain,xbar),np.subtract(ytrain,ybar)))/sum(np.multiply(np.subtract(xtrain,xbar),np.subtract(xtrain,xbar)))
    b0 = ybar - b1*xbar

    ypred = np.multiply(xtrain,b1)
    ypred = [x+b0 for x in ypred]

    sse=sum(np.multiply(np.subtract(ypred,ytrain),np.subtract(ypred,ytrain)))
    print("sse -> ",sse)


    plt.figure(i)
    plt.plot(xtrain,ytrain,label="original")
    plt.plot(xtrain,ypred,label="regression")
    plt.xlabel('time')
    plt.ylabel('No. of VMs')
    pylab.legend(loc='upper left')
    filename = '6a_'+str(i+1)+'.jpg'
    plt.savefig(filename)
    plt.show()
```
