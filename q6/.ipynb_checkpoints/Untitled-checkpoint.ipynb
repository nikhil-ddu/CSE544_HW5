```python
import numpy as np
f= open("q6.txt","r")
z=f.read().split("\n")
del z[-1]
#print(len(z))
ylist = [int(str.split("\t")[1]) for str in z]
xlist = list(range(1,577))
print(xlist, ylist)


```

```python
for i in range(4):
    xtrain = xlist [i*144:(i+1)*144]
    ytrain = ylist [i*144:(i+1)*144]
    xbar = np.mean(xtrain)
    ybar = np.mean(ytrain)

    b1 = sum(np.multiply(np.subtract(xtrain,xbar),np.subtract(ytrain,ybar)))/sum(np.multiply(np.subtract(xtrain,xbar),np.subtract(xtrain,xbar)))
    b0 = ybar - b1*xbar

    ypred = np.multiply(xtrain,b1)
    ypred = [x+b0 for x in ypred]

    sse=sum(np.multiply(np.subtract(ypred,ytrain),np.subtract(ypred,ytrain)))
    print("sse -> ",sse)


    plt.figure(i)
    plt.plot(xtrain,ytrain,label="original")
    plt.plot(xtrain,ypred,label="regression")
    plt.xlabel('time')
    plt.ylabel('No. of VMs')
    pylab.legend(loc='upper left')   
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import math

def linear_regression(x,y):
	print(x)
	print(y)
	x_bar = np.mean(x)
	y_bar = np.mean(y)
	num=0
	den =0
	for i in range(len(x)):
		num+=(x[i]-x_bar)*(y[i]-y_bar)
		den+=(x[i]-x_bar)**2
	beta1 = num/den
	beta0 = y_bar - (beta1*x_bar)
	regression_fit=[]
	for i in range(len(x)):
		regression_fit.append(beta0 + beta1*x[i])
	print("reg fit is")
	print(regression_fit)
	return regression_fit

Y=[]
X=[i for i in range(1,577)]
with open("q6.txt") as inputFile:
	for line in inputFile:
		if line.strip():
			Y.append(int("\t".join(line.split()[1:])))

Y = np.array_split(Y,4)
X = np.array_split(X,4)
reg_fit = []
reg_fit.append(linear_regression(X[0],Y[0]))
#reg_fit.append(linear_regression(X[1],Y[1]))
#reg_fit.append(linear_regression(X[2],Y[2]))
#reg_fit.append(linear_regression(X[3],Y[3]))
plt.plot(Y[0])
plt.plot(reg_fit[0])
#plt.plot(Y[1])
#plt.plot(reg_fit[1]) 
#plt.plot(Y[2])
#plt.plot(reg_fit[2]) 
#plt.plot(Y[3])
#plt.plot(reg_fit[3]) 
#plt.show()

	
```
