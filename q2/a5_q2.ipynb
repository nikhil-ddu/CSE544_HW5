```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
x=np.genfromtxt('A5_q2.dat',dtype="double",delimiter=",")
xlist=[]
ylist=[]
for i in range(x.shape[0]):
    xlist.append(x[i][0])
    ylist.append(x[i][1])
```

```python
sse=0
mape=0
ypred=[]

xtrain = xlist
xbar = np.mean(xtrain)
ybar = np.mean(ylist)

#print(xbar, ybar)

b1 = sum(np.multiply(np.subtract(xtrain,xbar),np.subtract(ylist,ybar)))/sum(np.multiply(np.subtract(xtrain,xbar),np.subtract(xtrain,xbar)))
b0 = ybar - b1*xbar
#print(b0)
#print(b1)

ypred = np.multiply(xtrain,b1)
ypred = [x+b0 for x in ypred]

#print(ypred)

mape=sum(np.divide(abs(np.subtract(ypred, ylist)), ylist))/len(ylist)
print("mape % -> ",mape*100 )

sse=sum(np.multiply(np.subtract(ypred,ylist),np.subtract(ypred,ylist)))
print("sse -> ",sse)

#print("mape is ",mape," %")
#print("sse is ",sse)
plt.figure(0)
plt.plot(xlist,ylist,label="original")
plt.plot(xlist,ypred,label="regression fit")
plt.xlabel('server CPU usage percentage(old X values)')
plt.ylabel('response time(Y)')
pylab.legend(loc='upper left')
plt.savefig('q2_a_1.jpg')
plt.show()
plt.figure(1)
plt.plot(xlist,ylist,label="original")
plt.plot(xlist,ypred,label="regression fit")
plt.xlabel('server CPU usage percentage(new X=using predictor(1) X )')
plt.ylabel('response time(Y)')
pylab.legend(loc='upper left')
plt.savefig('q2_a_2.jpg')


plt.show()
#print(ypred)
```
